# Task 1 – Hill Climbing
Use the Hill Climbing algorithm to find the maximum value of the function <br>
**f(x) = -x2 + 6x** <br>
**Range:** 0 ≤ x ≤ 6 <br>
**Instructions:** <br>
• Start from a random value of x. <br>
• Check the neighbor values (x+1 and x−1). <br>
• Move to the neighbor with the higher value of f(x). <br>
• Stop when no better neighbor exists. <br>
**Output should show:** <br>
• Initial value of x <br>
• Value of f(x) at each step <br>
• Final optimal value.

```python
import random

def f(x):
    return -x**2 + 6*x

def hill_climbing():
    current_x = random.randint(0, 6) #random value to start from
    print("Initial x:", current_x)
    print("f(x) =", f(current_x))
    while True:
        neighbors = []
        if current_x - 1 >= 0:
            neighbors.append(current_x - 1)
        if current_x + 1 <= 6:
            neighbors.append(current_x + 1)

        best_neighbor = current_x
        for n in neighbors:
            if f(n) > f(best_neighbor):
                best_neighbor = n
        if best_neighbor == current_x:
            print("\nFinal optimal value reached")
            print("Optimal x:", current_x)
            print("Maximum f(x):", f(current_x))
            break
        current_x = best_neighbor
        print("Move to x =", current_x, "f(x) =", f(current_x))

hill_climbing()
```

# Task 2 – Beam Search
Use Beam Search to reach the number 20 starting from 1. <br>
**Allowed operations:** <br>
• x → x + 2 <br>
• x → x + 3 <br>
• x → x × 2 <br>
**Heuristic function:** <br>
h(n) = |20 − current number| <br>
**Beam width:** k = 2 <br>
**Instructions:** <br>
• Start from number 1. <br>
• Generate the next possible numbers using the allowed operations. <br>
• At each level keep only the best 2 states with the lowest heuristic value. <br>
• Continue until the goal number 20 is reached. <br>
Output should show: <br>
• Numbers explored at each level <br>
• Final path to reach 20.

```python
import heapq

goal = 20
beam_width = 2
def heuristic(n):
    return abs(goal - n)

def successors(x):
    return [x + 2, x + 3, x * 2]

def beam_search(start):
    beam = [[start]]
    level = 0
    while beam:
        print(f"Level {level} explored:", [path[-1] for path in beam])
        candidates = []
        for path in beam:
            current = path[-1]
            if current == goal:
                return path
            for next_state in successors(current):
                if next_state <= 40:
                    new_path = path + [next_state]
                    h = heuristic(next_state)
                    candidates.append((h, new_path))

        best = heapq.nsmallest(beam_width, candidates, key=lambda x: x[0])
        beam = [path for h, path in best]
        level += 1
    return None

path = beam_search(1)
if path:
    print("\nFinal path to reach 20:")
    print(" → ".join(map(str, path)))
else:
    print("Goal not reached.")
```

# Task 3 – Genetic Algorithm
Use a Genetic Algorithm to maximize the function <br>
**f(x) = x2 + 2x** <br>
**Range:** 0 ≤ x ≤ 31 <br>
**Instructions:** <br>
• Use binary representation for chromosomes. <br>
• Generate a random initial population. <br>
• Calculate fitness using the function f(x). <br>
• Apply selection, crossover, and mutation. <br>
• Run the algorithm for 15 generations. <br>
**Output should show:** <br>
• Best chromosome <br>
• Best value of x <br>
• Best fitness value. <br>

```python
import random

def binary_to_decimal(chromosome):
    return int("".join(map(str, chromosome)), 2)

def fitness(chromosome):
    x = binary_to_decimal(chromosome)
    return x**2 + 2*x

def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    probabilities = [f/total_fitness for f in fitness_values]
    selected = random.choices(population, probabilities, k=2)
    return selected

def crossover(parent1, parent2):
    point = random.randint(1, len(parent1)-1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

def mutate(chromosome, mutation_rate):
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:
            chromosome[i] = 1 - chromosome[i]
    return chromosome

def generate_population(size, length):
    population = []
    for _ in range(size):
        chromosome = [random.randint(0,1) for _ in range(length)]
        population.append(chromosome)
    return population

def genetic_algorithm(population, generations, mutation_rate):
    for generation in range(generations):
        fitness_values = [fitness(ch) for ch in population]
        new_population = []
        while len(new_population) < len(population):
            parents = roulette_wheel_selection(population, fitness_values)
            child1, child2 = crossover(parents[0], parents[1])
            child1 = mutate(child1, mutation_rate)
            child2 = mutate(child2, mutation_rate)
            new_population.extend([child1, child2])
        population = new_population[:len(population)]
    best_chromosome = max(population, key=fitness)
    return best_chromosome

population_size = 6
chromosome_length = 5
generations = 15
mutation_rate = 0.01
initial_population = generate_population(population_size, chromosome_length)
best = genetic_algorithm(initial_population, generations, mutation_rate)
best_x = binary_to_decimal(best)
best_fitness = fitness(best)
print("Best Chromosome:", best)
print("Best value of x:", best_x)
print("Best fitness value:", best_fitness)
```
