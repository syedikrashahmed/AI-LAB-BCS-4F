## **1. BFS**


### **1.1 BFS (graph traversal)**

```python
def bfs(graph, start, goal):
  queue = []
  visited = []
  queue.append(start)
  visited.append(start)

  while queue:
    node = queue.pop(0)

    if node == goal:
      print("goal found")
      return

    for neighbor in graph[node]:
      if neighbor not in visited:
        visited.append(neighbor)
        queue.append(neighbor)
  print("goal not found")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
    }
bfs(graph, 'A', 'G')
```

### **1.2 BFS (matrix traversal)**

```python
from collections import deque
maze = [
    [1,1,0],
    [1,1,1,],
    [0,1,1]
]

rows = len(maze)
cols = len(maze[0])
start = (0,1)
goal = (2,0)

def bfs_matrix2(start,goal):
  visited = set([start])
  queue = deque([start])
  while queue:
    x,y = queue.popleft()

    if (x,y) == goal:
      print("goal found");
      return

    for dx,dy in [(0,1), (1,0), (0,-1), (-1,0)]:
      nx,ny = dx +x, dy +y
      if ((0<=nx<rows) and (0<=ny<cols) and (maze[nx][ny] == 1) and ((nx,ny) not in visited)):
        visited.add((nx,ny))
        queue.append((nx,ny))
  print("goal not found")

bfs_matrix2(start, goal)
```

### **1.3 BFS (goal based agent)**

```python
class Environmrnt:
  def  __init__(self):
    self.graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
    }

  def get_percept(self):
    return self.graph

class Goal_Based_Agent:
  def  __init__(self, start, goal):
    self.start = start
    self.goal = goal

  def formulate_goal(self, node):
    if node == self.goal:
      return True
    else:
      return False

  def bfs(self, graph):
    queue = []
    visited = []
    queue.append(self.start)
    visited.append(self.start)

    while queue:
      node = queue.pop(0)
      if (self.formulate_goal(node)):
        return "found"

      for neighbor in graph[node]:
        if neighbor not in visited:
          visited.append(neighbor)
          queue.append(neighbor)
    return "not found"

  def act(self, graph):
    return self.bfs(graph)

def agent_program(enviroment, agent):
  percept = enviroment.get_percept()
  action = agent.act(percept)
  if action == "found":
    print("goal was found in graph")
  else:
    print("goal was sadly not found")

environment = Environmrnt()
agent = Goal_Based_Agent('A', 'H')
agent_program(environment, agent)
```

## **2. DFS**

### **2.1 DFS (graph traversal)**

```python
def dfs(graph, start, goal):
  visited = []
  stack = []
  stack.append(start)
  visited.append(start)

  while stack:
    node = stack.pop(-1)
    if node == goal:
      print("yay")
      return

    for neighbor in reversed(graph[node]):
      if (neighbor not in visited):
        visited.append(neighbor)
        stack.append(neighbor)
  print("not found")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': [],
    'Z': []
    }
dfs(graph, 'A', 'G')
```

## **3. DLS**

```python
def dfs_for_dls(node, goal, graph, depth, depth_limit, visited):
  if depth > depth_limit:
    return None

  visited.append(node)
  if node == goal:
    print("goal found")
    return list(visited)

  for neighbor in graph[node]:
    if neighbor not in visited:
      path = dfs_for_dls(neighbor, goal, graph, depth + 1, depth_limit, visited)
      if path:
        return path
  visited.pop()
  return None

def dls(start, goal, graph, depth_limit):
  visited = []
  result = dfs_for_dls(start, goal, graph, 0, depth_limit, visited)
  if result:
    print(result)
  else:
    print("sucks to be u")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': [],
    'Z': []
}
dls('A', 'G', graph, 4)
```

## **4. IDS**

```python
#same as dls ka code bas ids ka code is different
def dfs(node, goal, graph, depth, depth_limit, visited):
    if depth > depth_limit:
        return None

    visited.append(node)
    if node == goal:
        return list(visited)

    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            path = dfs(neighbor, goal, graph, depth + 1, depth_limit, visited)
            if path:
                return path
    visited.pop()
    return None


def dls(graph, start, goal, depth_limit):
    visited = []
    result = dfs(start, goal, graph, 0, depth_limit, visited)
    if result is None:
        return None
    return result

def iterative_deepening(graph, start, goal, max_depth):
    for depth in range(max_depth + 1):
        print(f"Searching with depth limit: {depth}")
        if dls(graph, start, goal, depth) is not None:
            print(f"Goal {goal} found at depth {depth}")
            return
    print("Goal not found within depth limit")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': [],
    'Z': []
}
iterative_deepening(graph, 'A', 'I', 3)
```

## **5. UCS**

```python
import heapq
def ucs(start, goal, graph):
  frontier = [(0, start)]
  cost_so_far = {start: 0}
  came_from = {start: None}
  while frontier:
    curr_cost, curr_node = heapq.heappop(frontier)
    if curr_cost > cost_so_far[curr_node]:
      continue
    if curr_node == goal:
      path = []
      while curr_node is not None:
        path.append(curr_node)
        curr_node = came_from[curr_node]
      path.reverse()
      print(f"cost: {curr_cost} and path: {path}")
      return

    for neighbor, cost in graph[curr_node].items():
      new_cost = cost + curr_cost
      if (neighbor not in cost_so_far) or (new_cost < cost_so_far[neighbor]):
        cost_so_far[neighbor] = new_cost
        came_from[neighbor] = curr_node
        heapq.heappush(frontier, (new_cost, neighbor))
  print("not found goal")

graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}
ucs('A', 'G', graph)
```

## **7. A Star**

```python
import heapq
def a_star(start, goal, graph, heuristic):
  frontier = [(heuristic[start] + 0, start)]
  visited = []
  cost_so_far = {start: 0}
  came_from = {start: None}
  while frontier:
    curr_f, curr_node = heapq.heappop(frontier)

    if curr_node in visited:
      continue

    visited.append(curr_node)

    if curr_node == goal:
      path = []
      while curr_node is not None:
        path.append(curr_node)
        curr_node = came_from[curr_node]
      path.reverse()
      print(f"yoo found, path: {path}")
      return True

    for neighbor, cost in graph[curr_node].items():
      new_g = cost_so_far[curr_node] + cost
      new_f = new_g + heuristic[neighbor]
      if neighbor not in cost_so_far or new_g < cost_so_far[neighbor]:
        cost_so_far[neighbor] = new_g
        came_from[neighbor] = curr_node
        heapq.heappush(frontier, (new_f, neighbor))
  print("Not found shad")


graph = {
    'A': {'B': 4, 'C': 3},
    'B': {'E': 12, 'F': 5},
    'C': {'D': 7, 'E': 10},
    'D': {'E': 2},
    'E': {'G': 5},
    'F': {'G': 16},
    'G': {},
}

heuristic = {'A': 14, 'B': 12, 'C': 11, 'D': 6, 'E': 4, 'F': 11, 'G': 0}
a_star('A', 'G', graph, heuristic)
```

```python
import random

# convert binary to decimal
def to_decimal(ind):
    return int(''.join(map(str, ind)), 2)

# fitness function
def fitness(ind):
    x = to_decimal(ind)
    return x*x - 1

# create individual
def create_individual():
    return [random.randint(0,1) for _ in range(5)]

# initialize population
def init_population(size):
    return [create_individual() for _ in range(size)]

# selection (tournament)
def selection(pop):
    a, b, c = random.sample(pop, 3)
    return max([a,b,c], key=fitness)

# crossover
def crossover(p1, p2):
    point = random.randint(1, len(p1)-1)
    return p1[:point]+p2[point:], p2[:point]+p1[point:]

# mutation
def mutate(ind):
    if random.random() < 0.1:
        i = random.randint(0, len(ind)-1)
        ind[i] = 1 - ind[i]
    return ind

# main GA
def GA():
    pop = init_population(6)

    for _ in range(20):
        new_pop = []
        for _ in range(3):
            p1 = selection(pop)
            p2 = selection(pop)
            c1, c2 = crossover(p1, p2)
            new_pop.append(mutate(c1))
            new_pop.append(mutate(c2))
        pop = new_pop

    best = max(pop, key=fitness)
    print("Best:", best, "x=", to_decimal(best), "fitness=", fitness(best))

GA()
```
