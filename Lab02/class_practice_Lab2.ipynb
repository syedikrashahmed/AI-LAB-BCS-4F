# **Practice Task # 1**
Create a program to control smart lights in a house. Each light has a room name and a
status (ON/OFF). The system should allow:
1. Creating multiple light objects for different rooms.
2. Turning lights ON or OFF.
3. Displaying the current status of each light.

```python
class Light:
    def __init__(self, roomName, status):
        self.roomName = roomName
        self.status = status
    def switch(self):
        if self.status == "ON":
            self.status = "OFF"
        else:
            self.status = "ON"
    def display(self):
        print(self.roomName, self.status)

l1 = Light("Bedroom", "ON")
l1.display()
l1.switch()
l1.display()
l2 = Light("Kitchen", "OFF")
l2.display()

```

# **Practice Task # 2**
You are designing a University Staff Management System. The system needs to keep
track of different staff members: Teachers, Administrative Staff, and Research
Assistants. All staff share common attributes like name, staff_id, and department, but
each type has its own specific properties and behaviors. Teachers have courses and
salary and can teach courses, Admin Staff have role and working_hours and perform
tasks, while Research Assistants have research_topic and stipend and work on research.
Your task is to create a base class Staff and derive classes for each staff type, instantiate
objects, and call their respective methods to simulate daily activities. O

```python
class Staff:
  def __init__(self, name, staff_id, department):
    self.name = name
    self.staff_id = staff_id
    self.department = department

  def display(self):
    print(self.name, self.staff_id, self.department)

class Teacher(Staff):
    def __init__(self, name, staff_id, department, courses, salary):
      super().__init__(name, staff_id, department)
      self.courses = courses
      self.salary = salary

    def work(self):
      print(self.name, " is teaching ", self.courses)

class Admin(Staff):
    def __init__(self, name, staff_id, department, role, working_hours):
      super().__init__(name, staff_id, department)
      self.role = role
      self.working_hours = working_hours

    def work(self):
      print(self.name, " is doing ", self.role)

class ResearchAssistant(Staff):
    def __init__(self, name, staff_id, department, research_topic, stipend):
      super().__init__(name, staff_id, department)
      self.research_topic = research_topic
      self.stipend = stipend

    def work(self):
      print(self.name, " is researching ", self.research_topic)

teacher = Teacher("ikrash", 998, "CS", "PF", 500)
teacher.display()
teacher.work()
admin = Admin("babar", 56, "PCB", "batting", 5)
admin.display()
admin.work()
researchAssistant = ResearchAssistant("abdullah", 55, "PCB", "batting", 5)
researchAssistant.display()
researchAssistant.work()
```

# **Practice Task # 3**
You are designing a simple banking system where each customer has a bank account. The
account balance should be private and not directly accessible from outside the class. Your task
is to create a BankAccount class with a private attribute __balance and public methods to
deposit, withdraw, and get_balance. Create a few account objects and use these methods to

```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount

  def withdraw(self, amount):
    self.__balance -= amount

  def get_balance(self):
    return self.__balance

a1 = BankAccount(5000)
print("balance is ", a1.get_balance())
a1.deposit(50)
print("balance after deposit ", a1.get_balance())
a1.withdraw(4000)
print("balance after withdraw ", a1.get_balance())
```
