Question 1: Student Grade System
Question:
Write a Python program that takes a student’s name and marks as input and prints the grade
according to the following criteria:
• Marks ≥ 85 → Grade A
• Marks ≥ 70 → Grade B
• Marks ≥ 50 → Grade C
• Marks < 50 → Fail

```python
name = input("Enter your name: ")
marks = int(input("Enter your marks: "))
print("Student Name: ", name)
if marks >= 85:
    print("Marks: ", marks, "\nGrade: A")
elif marks >= 70:
    print("Marks: ", marks, "\nGrade: B")
elif marks >= 50:
    print("Marks: ", marks, "\nGrade: C")
else:
    print("Marks: ", marks, "\nGrade: Fail")
```

Question 2: Even Number Counter
Question:
Write a Python program that takes a number n from the user and prints all even numbers from 1
to n. Also display the total count of even numbers.

```python
n = int(input("Enter a number: "))
total = 0
print("Even numbers: ")
for i in range(1, n + 1):
    if i % 2 == 0:
        print(i)
        total += 1
print("\nTotal even numbers: ", total)
```

Question 3: Student Record Using Dictionary
Question:
Write a Python program that stores the names and marks of three students in a dictionary and then displays all student records.

```python
dict = {}
for i in range(3):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    dict[name] = marks

print("\nStudent Records: ")
for i in dict:
    print(i, ": ", dict[i])
```

Question 4: Menu-Driven Calculator
Question:
Write a Python program using a while loop that shows the following menu repeatedly until the
user exits:
1. Add two numbers
2. Subtract two numbers
3. Exit

```python
choice = 0
while choice != 3:
  choice = int(input("Enter 1 to add two numbers \n2 to subtract two numbers \n3 to quit \nChoice: "))
  if choice == 1:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    print("Sum: ", num1 + num2)
  elif choice == 2:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    print("Difference: ", num1 - num2)
  elif choice == 3:
    print("Goodbye!")
  else:
    print("Invalid choice")
```

Question 5: Average Marks Using Function
Question:
Write a Python function named calculate_average() that takes a list of marks and returns the
average marks.

```python
total = 0
marks = []
n = int(input("Enter number of marks: "))
for i in range(n):
  marks.append(int(input("Enter marks: ")))

for i in range(len(marks)):
    marks[i] = int(marks[i])
    total += marks[i]

average = total / len(marks)
print("Average Marks: ", average)
```
